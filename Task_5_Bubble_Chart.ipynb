Bubble Chart: Analyze the relationship between app size (in MB) and average rating, with bubble size representing the number of installs. Include only apps with a rating higher than 3.5 in the "Games" category and installs more than 50k. This graph should be displayed only between 5 PM IST and 7 PM IST.

```python
import pandas as pd
data = pd.read_csv('cleaned_playstore_data.csv')
data.head()
```

```python

import altair as alt
import pandas as pd
data['Installs'] = pd.to_numeric(data['Installs'].astype(str).str.replace(r'[,+]', '', regex=True), errors='coerce')

# Filter for relevant apps
filtered_data = data[
    (data['Category'] == 'GAME') &
    (data['Rating'] > 3.5) &
    (data['Installs'] > 50000)
]
# Create the bubble chart
chart = alt.Chart(filtered_data).mark_circle().encode(
    x='Size:Q',
    y='Rating:Q',
    size='Installs:Q',
    tooltip=['App', 'Rating', 'Size', 'Installs']
)

chart
```

```python

import datetime

now = datetime.datetime.now()
current_hour = now.hour

current_hour_ist = (current_hour + 5) % 24 + 0.5

if 17 <= current_hour_ist < 19:  # Check if it's between 5 PM and 7 PM IST
  chart.display()
else:
  # for my own verification
 # chart.display()
 print("The bubble chart is only available between 5 PM IST and 7 PM IST.")
```
