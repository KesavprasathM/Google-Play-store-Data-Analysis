```python
data = pd.read_csv('cleaned_playstore_data.csv')
data['Installs'] = data['Installs'].astype(str).str.replace(',', '', regex=True).str.replace(r'\+', '', regex=True)
data['Installs'] = pd.to_numeric(data['Installs'], errors='coerce')
data.dropna(subset=['Installs'], inplace=True)

# Filter categories
excluded_categories = ['A', 'C', 'G', 'S']
data = data[~data['Category'].str.startswith(tuple(excluded_categories))]

# top 5 categories by installs
top_categories = data.groupby('Category')['Installs'].sum().nlargest(5).index
data = data[data['Category'].isin(top_categories)]
data['Installs'] = pd.to_numeric(data['Installs'], errors='coerce')

# display the choropleth map
def create_choropleth():
    fig = px.choropleth(data_frame=data, locations="App", locationmode='country names',
                        color="Installs", hover_name="App", hover_data=["Category", "Installs"],
                        range_color=[0, data['Installs'].max()],
                        title="Global Installs by App (Top 5 Categories)")

    # Highlight installs > 1 million
    fig.update_traces(marker=dict(line=dict(width=1, color='DarkSlateGrey')), selector=dict(mode='markers'))
    fig.add_shape(type='rect', x0=data.index.min(), y0=1e6, x1=data.index.max(), y1=data['Installs'].max(),
                  line=dict(color='black', width=1), fillcolor='lightgray', opacity=0.5)


    fig.show()
ist_tz = pytz.timezone('Asia/Kolkata')
now_ist = datetime.now(ist_tz)
current_hour_ist = now_ist.hour

if 18 <= current_hour_ist <= 20:  # 6 PM to 8 PM IST
  create_choropleth()
  # for my own verfication does the map is working or not even you can also use this !
#else:
    #create_choropleth()
else:
  print("Interactive map is only available between 6 PM and 8 PM IST.")
```
