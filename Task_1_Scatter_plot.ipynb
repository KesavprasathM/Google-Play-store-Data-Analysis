```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data= pd.read_csv('Play Store Data.csv')  # Load the data set
data.head()
```

```python

 #Replace 'M' with '000000' and 'K' with '000' in the 'Size' column
data['Size'] = data['Size'].str.replace('M', '000000', regex=False)
data['Size'] = data['Size'].str.replace('k', '000', regex=False)
```

```python
data.head()
```

```python
data.dtypes
```

```python
# Drop rows with missing 'Rating' values
data = data.dropna(subset=['Rating'])

# Fill missing values in each column with the mode of that column
for column in data.columns:
    data[column].fillna(data[column].mode()[0], inplace=True)

# Remove duplicate rows
data.drop_duplicates(inplace=True)

# Filter rows where 'Rating' is less than or equal to 5
data = data[data['Rating'] <= 5]

# Drop rows with missing 'Translated_Review' values in reviews_df
# Reviews_df.dropna(subset=['Translated_Review'], inplace=True)
```

```python
data.dtypes
```

```python
#Convert the Installs columns to numeric by removing commas and +
data['Installs']=data['Installs'].astype(str).str.replace(',','').str.replace('+','').astype(int)

#Convert Price column to numeric after removing $
data['Price']=data['Price'].astype(str).str.replace('$','').astype(float)
```

```python
# Convert the Installs columns to numeric by removing commas and +
data['Installs'] = data['Installs'].astype(str).str.replace(',', '').str.replace('+', '')
# Convert to numeric, handling errors and invalid values
data['Installs'] = pd.to_numeric(data['Installs'], errors='coerce').fillna(0).astype(int)

# Convert Price column to numeric after removing $
data['Price'] = data['Price'].astype(str).str.replace('$', '').astype(float)
```

```python
data.dtypes
```

```python
#Revenue column
data['Revenue']=data['Price']*data['Installs']
```

```python
data.head()
```

```python
# TO CREAT A SCATTER PLOT TO VISUALIZE THE RELATIONSHIP BETWEEN REVENUE VS THE NUMBER OF INSTALLS FOR PAID APPS ONLY
# ADD A TREND LINE TO SHOW THE CORRELATION AND COLOR-CODE THE POINTS BASED ON APP CATEGORIES

import pandas as pd
import plotly.express as px

# Convert 'Installs' and 'Price' columns to numeric, handling errors
data['Installs'] = data['Installs'].astype(str).str.replace(r'[+,]', '', regex=True)
data['Installs'] = pd.to_numeric(data['Installs'], errors='coerce')
data['Price'] = data['Price'].astype(str).str.replace(r'[$]', '', regex=True)
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')

# Filter for paid apps
paid_apps = data[data['Type'] == 'Paid']

# Create the scatter plot
fig = px.scatter(
    paid_apps,
    x='Installs',
    y='Price',
    color='Category',  # Color-code TO category
    trendline='ols',  # Add a trendline
    title=' REVENUE VS INSTALLS FOR PAID APPS',
    labels={'Installs': 'NUMBER OF INSTALLS', 'Price': 'REVENUE'},
    width=900,  # width of plot
    height=600   # height of plot
)

fig.show()
```

```python


# Create a copy of the original dataframe to store the cleaned data
cleaned_data = data.copy()

# Convert 'Reviews' column to numeric, coercing errors to NaN
cleaned_data['Reviews'] = pd.to_numeric(cleaned_data['Reviews'], errors='coerce')

# Convert 'Size' column to numeric, handling various formats
cleaned_data['Size'] = cleaned_data['Size'].str.replace('M', 'e6')
cleaned_data['Size'] = cleaned_data['Size'].str.replace('k', 'e3')
cleaned_data['Size'] = cleaned_data['Size'].str.replace('Varies with device', 'NaN') # replace 'varies with device' with NaN
cleaned_data['Size'] = pd.to_numeric(cleaned_data['Size'], errors='coerce')


# Remove rows with NaN values in crucial columns for analysis
cleaned_data.dropna(subset = ['Reviews', 'Size'], inplace=True)

# Removing '$' sign from 'Price' column and converting it to numeric
cleaned_data['Price'] = cleaned_data['Price'].apply(lambda x : str(x).replace('$','') if '$' in str(x) else str(x))
cleaned_data['Price'] = pd.to_numeric(cleaned_data['Price'], errors = 'coerce')
#Handling the Installs column
cleaned_data['Installs'] = cleaned_data['Installs'].astype(str).str.replace('+','')
cleaned_data['Installs'] = cleaned_data['Installs'].astype(str).str.replace(',','')
cleaned_data['Installs'] = pd.to_numeric(cleaned_data['Installs'])
#Drop rows with null values after cleaning
cleaned_data.dropna(subset = ['Price', 'Installs'], inplace = True)

# Print some info of cleaned data
cleaned_data.info()
```

```python


cleaned_data.to_csv('cleaned_playstore_data.csv', index=False)
```
